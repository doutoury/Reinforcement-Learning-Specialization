Different types of models <br><br>

  1. __Sample model__ <br>
  produces an actual outcome drawn from some underlying porbabilies. 
  > Ex. a sample model for flipping a coin can generate a random sequence of heads and tails.

  <br>
  
  2. __Distribution model__ <br>
  Completely specifies the likelyhood or probability of every outcome. <br>

### __Distribution model example $\quad : \;$ Rolling 12 dice__ <br><br>


Modelling __the complete joint distribution__ of rolling 12 dice is much more work. <br>
We must consider __every possible outcome__ of 12 dice, and assign a probability to each outcome. <br><br>

With three dice, we have 216 outcomes <br>

With twelve dice, we have 46656 outcomes to consider ...

But it can be useful <br>
__" a Distribution model produces tha exact probability of an outcomes ! "__ <br><br>

$\rightarrow \quad$ we can __compute the expected outcome__ dicrectly <br>

#### Sample model <br><br>

$\quad$ pros. <br>
  - require __less memory__ <br>

$\quad$ cons. <br>
  - the Sample models can __only approximate__ this __expected outcome__ <br>
  by averaging many samples together <br>

<br><br>


#### Distribution model <br><br>

$\quad$ pros. <br>
  - cen be used to __compute__ the __exact expected outcome__ <br>
  by summing over all (possible) outcomes weighted by their probabilities <br>
  - Knowing the __exact probabilties__ also has __the flexibility of assessing risk__. <br>
  >For example. <br>
  >When a doctor is prescribing medicine, <br>
  >they would consider many possible side effects and how likely they are to occur.

### __Summary__ <br><br>

  - Sample models can be more compact than distribution models <br><br>

  - The explicit probabilities stored by Distribution models <br>
  can be used to exactly identify the expected outcome
